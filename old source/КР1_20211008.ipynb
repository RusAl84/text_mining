```python
!pip install pymorphy2
import pymorphy2
morph = pymorphy2.MorphAnalyzer()

words = []
for item in texts.split():
  #print(item)
  words.append(item)

#words = texts.split()

#words = ['грустно', 'зависимость', 'хорошему', 'приводит', 'альтернатив']
lemTextsList=[]
lemTexts=""
for word in words:
    p = morph.parse(word)[0]
    #print(p.normal_form)
    lemTextsList.append(p.normal_form)
    lemTexts = lemTexts + " " + p.normal_form
print(lemTextsList)
print(lemTexts)
print(texts)
```
